```python
import numpy as np
import random
```

```python

# Define parameters
num_customers = 32
num_vehicles = 4
capacity = 100
num_epochs = 1000
learning_rate = 0.1
discount_factor = 0.9
epsilon = 0.1

# Define customer data
customer_data = {
    1: (42, 68, 0), 2: (77, 97, 5), 3: (28, 64, 23), 4: (77, 39, 14),
    5: (32, 33, 13), 6: (32, 8, 8), 7: (42, 92, 18), 8: (8, 3, 19),
    9: (7, 14, 10), 10: (82, 17, 18), 11: (48, 13, 20), 12: (53, 82, 5),
    13: (39, 27, 9), 14: (7, 24, 23), 15: (67, 98, 9), 16: (54, 52, 18),
    17: (72, 43, 10), 18: (73, 3, 24), 19: (59, 77, 13), 20: (58, 97, 14),
    21: (23, 43, 8), 22: (68, 98, 10), 23: (47, 62, 19), 24: (52, 72, 14),
    25: (32, 88, 13), 26: (39, 7, 14), 27: (17, 8, 2), 28: (38, 7, 23),
    29: (58, 74, 15), 30: (82, 67, 8), 31: (42, 7, 20), 32: (68, 82, 24),
}


# Calculate distances between customers
distances = np.zeros((num_customers + 1, num_customers + 1))
for i in range(1, num_customers + 1):
    for j in range(1, num_customers + 1):
        if i != j:
            distances[i][j] = np.sqrt((customer_data[i][0] - customer_data[j][0]) ** 2 +
                                      (customer_data[i][1] - customer_data[j][1]) ** 2)

# Q-learning function
def q_learning():
    Q = np.zeros((num_customers + 1, num_customers + 1))  # Q-table
    all_routes = []

    for _ in range(10):  # Generate 10 different routes
        current_capacity = capacity
        current_location = 0  # Start at depot
        unvisited_customers = list(range(1, num_customers + 1))
        route = [0]

        while len(route) < num_customers + 1:
            if random.uniform(0, 1) < epsilon:
                next_customer = random.choice(unvisited_customers)
            else:
                next_customer = np.argmax(Q[current_location])

            if next_customer not in unvisited_customers:
                next_customer = random.choice(unvisited_customers)

            next_location = next_customer
            if next_location != 0:  # Exclude depot
                current_capacity -= customer_data[next_location][2]
            route.append(next_location)
            current_location = next_location

            if next_customer in unvisited_customers:
                unvisited_customers.remove(next_customer)

            # Q-table update
            reward = -distances[current_location][next_location]
            Q[current_location][next_location] += learning_rate * (
                reward + discount_factor * np.max(Q[next_location]) - Q[current_location][next_location]
            )

        route.append(0)  # End at depot

        # Check if all customers have been visited
        if len(unvisited_customers) > 0:
            missing_customer = random.choice(unvisited_customers)
            route.insert(random.randint(1, num_customers), missing_customer)

        all_routes.append(route)

    return all_routes

# Main function
if _name_ == "_main_":
    routes = q_learning()
    for i, route in enumerate(routes):
        print(f"Route {i + 1}: {route}")
```
